### 실손보험에서 급여 본인부담금에 대한 보험사의 보험금 지출 차이 예측

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/jieunpark/문서/25_2_EDA/merged_preprocessed_with_22.csv", encoding='euc-kr', index_col=0)
df
```

```python
df['연령'].unique(), df['298 질병분류'].unique()
```

```python
# 외래 진료비에 대한 건강보험 평균 본인부담률 계산
df_외래 = df[df['진료형태'] == '외래'].copy()
df_외래 = df_외래[~df_외래['연령'].isin(['계']) & ~df_외래['298 질병분류'].isin(['계', '기타'])]
df_외래
```

```python
df['22대 질병분류'].unique()
```

```python
a = df_외래['진료비'].sum()
a_임신출산 = df_외래[df_외래['22대 질병분류'] == "15"]['진료비'].sum()
b = df_외래['급여비'].sum()
b_임신출산 = df_외래[df_외래['22대 질병분류'] == "15"]['급여비'].sum()

value = (a-b)/a
formatted = f"{value * 100:.2f}%"

print("외래 급여에 대한 건강보험 평균 본인부담률:", formatted)
```

```python
v4세대_임신출산고려X = (a-b) * (1-0.2) * 1000
v4세대_임신출산고려 = ((a-a_임신출산)-(b-b_임신출산)) * (1-0.2) * 1000
v5세대 = (a-b) * (1-0.285) * 1000

print(f"4세대 기준 2023년 급여 본인부담금에 대한 실손보험 보험금 지출 예측 (임신출산 고려X): 약 {round(v4세대_임신출산고려X):,.0f}(원)")
print(f"4세대 기준 2023년 급여 본인부담금에 대한 실손보험 보험금 지출 예측: 약 {round(v4세대_임신출산고려):,.0f}(원)")
print(f"5세대 기준 2023년 급여 본인부담금에 대한 실손보험 보험금 지출 예측: 약 {round(v5세대):,.0f}(원)")
```

```python
# cf. 입원 진료비에 대한 건강보험 평균 본인부담률 (20% 언저리여야함)
df_입원 = df[df['진료형태'] == '입원'].copy()
df_입원 = df_입원[~df_입원['연령'].isin(['계']) & ~df_입원['298 질병분류'].isin(['계', '기타'])]

a = df_입원['진료비'].sum()
b = df_입원['급여비'].sum()

value = (a-b)/a
formatted = f"{value * 100:.2f}%"

print("입원 급여에 대한 건강보험 평균 본인부담률:", formatted)
```
